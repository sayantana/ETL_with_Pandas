```python

my_list = [y for y in os.listdir(DATA_DIR) if y.endswith('.csv') ]

finaldb = []
for i in my_list:
    df2 = pd.read_csv(os.path.join(DATA_DIR, i))
    year_val = i.replace(".csv", "")
    df2['year']=year_val
    df2['filename']=i
    finaldb.append(df2)


df3 = pd.concat(finaldb)

df3

# df3.concat(finaldb)

   
```
