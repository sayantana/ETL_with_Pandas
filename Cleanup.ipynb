```python
import os
import pandas as pd
```

```python
cwd = os.getcwd()
DATA_DIR = os.path.join(cwd, 'data')
CACHE_DIR = os.path.join (cwd, 'cache')
working_file = os.path.join(CACHE_DIR, 'output_1.csv')
```

```python
df = pd.read_csv(working_file)
```

```python
df['Rank']=-1
```

```python
df.head()
```

```python
df['Domestic %']=df['%']
df['Foreign %']=df['%.1']
df.head()
```

```python
df.drop(['%', '%.1'], axis=1, inplace=True)
df.head()
```

```python


#df['Worldwide']=df['Worldwide'].replace('$', '').replace(',', '')
col_list=['Worldwide', 'Domestic', 'Foreign']

def col_to_int(str):
    str = str.replace('$', '').replace(',', '')
    try:
        int_str = int(str)
    except:
        int_str = 0
    return int_str

    
    

def clean_column(row):
    for i in col_list:
        row[i] = col_to_int(row[i])
    return row

df2_cleaned = df.apply(clean_column, axis=1)
df2_cleaned.head()
```

```python
df2_cleaned.sort_values (by ='Worldwide', ascending=False, inplace=True)
df2_cleaned.head()
```

```python
df2_cleaned.reset_index(inplace=True, drop=True)
df2_cleaned.head()
```

```python
df2_cleaned['Rank']=df2_cleaned.index + 1
df2_cleaned.head()
```

```python
df2_cleaned['Domestic %']= df2_cleaned['Domestic']/df2_cleaned['Worldwide']*100
df2_cleaned['Foreign %']= df2_cleaned['Foreign']/df2_cleaned['Worldwide']*100
df2_cleaned.head()
```
